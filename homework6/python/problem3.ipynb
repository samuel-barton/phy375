```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
rcParams['figure.figsize'] = (10, 10)
rcParams['font.size'] = 20

A = np.radians(30) # the upper angle of the isocoles triangle
n = 1.5 # index of refraction of the prism
```

```python
def delta(theta_1):
    theta_1 = np.radians(theta_1)
    return np.degrees(theta_1 + np.arcsin(n*np.sin(A - np.arcsin(np.sin(theta_1/n)))) - A)
```

```python
def tParallel(theta_i,n_i,n_t):
    theta_t = np.arcsin(n_i*np.sin(theta_i)/n_t)
    numerator = 2*np.sin(theta_t)*np.cos(theta_i)
    denominator = np.sin(theta_i + theta_t)
    
    return numerator/denominator
```

```python
def tPerp(theta_i,n_i,n_t):
    theta_t = np.arcsin(n_i*np.sin(theta_i)/n_t)
    numerator = 2*np.sin(theta_t)*np.cos(theta_i)
    denominator = np.sin(theta_i + theta_t)*np.cos(theta_i - theta_t)

    return numerator/denominator
```

```python
def TParallel(theta_1):
    theta_1 = np.radians(theta_1)
    theta_2 = np.arcsin(np.sin(theta_1/n))
    theta_3 = A - theta_2

    T_parallel_1 = tParallel(theta_1,1,n)**2
    T_parallel_2 = tParallel(theta_3,n,1)**2
    
    return T_parallel_1*T_parallel_2
```

```python
def TPerpendicular(theta_1):
    theta_1 = np.radians(theta_1)
    theta_2 = np.arcsin(np.sin(theta_1/n))
    theta_3 = A - theta_2
    T_per_1 = tPerp(theta_1,1,n)**2
    T_per_2 = tPerp(theta_3,n,1)**2
    
    return T_per_1*T_per_2
```

```python
thetas = np.linspace(0,90, 10000)

plt.plot(thetas, delta(thetas), "r.")
plt.ylabel(r'$\delta(\theta_1)$')
plt.xlabel(r'$\theta_1$')
plt.savefig('../latex/delta-vs-theta-1.png', bbox_inches='tight')
plt.show()
plt.plot(thetas,TPerpendicular(thetas),"r.")
plt.ylabel(r'$\mathrm{T}_\bot(\theta_1)$')
plt.xlabel(r'$\theta_1$')
plt.axvline(0.0,color='black', linestyle="--")
plt.axhline(0.0,color='black', linestyle="--")
plt.savefig('../latex/T-per-vs-theta-1.png', bbox_inches='tight')
plt.show()
plt.plot(thetas,TParallel(thetas),"r.")
plt.ylabel(r'$\mathrm{T}_\parallel(\theta_1)$')
plt.xlabel(r'$\theta_1$')
plt.axvline(0.0,color='black', linestyle="--")
plt.axhline(0.0,color='black', linestyle="--")
plt.savefig('../latex/T-par-vs-theta-1.png', bbox_inches='tight')
plt.show()
plt.plot(thetas,TPerpendicular(thetas),"r.")
plt.ylabel(r'$\mathrm{T}_\bot(\theta_1)$')
plt.xlabel(r'$\theta_1$')
plt.axvline(0.0,color='black', linestyle="--")
plt.axhline(0.0,color='black', linestyle="--")
plt.savefig('../latex/T-per-vs-theta-1.png', bbox_inches='tight')
plt.show()
```
