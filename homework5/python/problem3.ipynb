```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from pylab import rcParams
rcParams['figure.figsize'] = (10, 10)
rcParams['font.size'] = 20
```

```python
n_incident = 1.0
n_transmitted = 1.33
```

```python
def thetaT(n_i,n_t, theta_i):
    return np.arcsin((n_i*np.sin(theta_i))/n_t)
```

```python
def rParallel(n_i, n_t, theta_i):
    theta_t = thetaT(n_i, n_t, theta_i)
    return np.tan(theta_i - theta_t)/np.tan(theta_i + theta_t)
```

```python
def brewsters(n_i, n_t):
    return 360/(2*np.pi)*np.arctan(n_t/n_i)
```

```python
thetas = np.linspace(0,np.pi/2, 1000)
theta_deg = thetas*(360/(2*np.pi))
r_par =  rParallel(n_incident, n_transmitted, thetas)

b = brewsters(n_incident, n_transmitted)

plt.axvline(b,color='red', linestyle="--", label="x-intercept")
plt.axhline(0.0,color='black', linestyle="--", label="zero")
plt.ylabel(r'$\mathrm{r}_\parallel$')
plt.xlabel(r'$\theta_i$')
plt.plot(theta_deg, r_par, "b-", label=r'$r_\parallel \;vs\; \theta_i$')
plt.annotate(r'$\theta_B$ = '+str(np.round(b,3))+ r'$^\circ$', 
             xy=(b,0.0), 
             xytext=(b + 1,0.02))
plt.legend()
plt.savefig('r-par-1.png', bbox_inches='tight')
plt.show()
```
